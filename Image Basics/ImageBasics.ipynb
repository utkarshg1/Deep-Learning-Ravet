# Image Basics 

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Reading the image in BGR
img = cv2.imread('CuteCat.jpeg')
img
```

```python
img.shape
```

```python
# Plot this image using matplotlib
plt.imshow(img)
```

### Converting image from BGR to RGB

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

### Grayscale image

```python
img3 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img3.shape
```

```python
plt.imshow(img3, cmap='gray')
```

### Splitting the channels b,g,r from image

```python
b, g, r = cv2.split(img)
```

```python
b
```

```python
g
```

```python
r
```

```python
# Viewing individual channels
plt.imshow(b,cmap='Blues')
```

```python
# Green channel
plt.imshow(g,cmap='Greens')
```

```python
# Red Channel
plt.imshow(r, cmap='Reds')
```

### How pixel values are stored in image

```python
B = cv2.imread('Alphabets.jpg',cv2.IMREAD_GRAYSCALE)
B
```

```python
B.shape
```

```python
plt.imshow(B,cmap='gray')
```

### Visualize pixel values

```python
import seaborn as sns
plt.figure(figsize=(15,15))
sns.heatmap(B, annot=True, fmt='d')
plt.show()
```
