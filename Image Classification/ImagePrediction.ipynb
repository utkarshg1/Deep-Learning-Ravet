# Prediction of Handwritten digit

```python
# Load the model
from keras.models import load_model
```

```python
model = load_model('Handwritten_model.keras')
```

```python
model.summary()
```

### Create a function to load image and predict data

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
def predict_image(path,model):
    # Import and display image
    img = cv2.imread(path,cv2.IMREAD_GRAYSCALE)
    plt.imshow(img, cmap='gray')
    plt.show()

    # Apply scaling to image
    img_scl = img/255

    # Get the image array (28,28,1)
    img_scl = np.array([img_scl])

    # Predict the probability
    prob = model.predict(img_scl)

    # Class prediction
    pred = np.argmax(prob,axis=-1)[0]

    # Class prob
    max_prob = prob.max()

    print(f'Prediction is {pred} with probability of {max_prob:.4f}')

    return pred, max_prob
```

```python
predict_image('Digit2.jpg',model)
```

```python
predict_image('Digit7.jpg',model)
```
