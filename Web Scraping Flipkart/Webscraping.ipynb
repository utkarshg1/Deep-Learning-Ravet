### Create a function to fetch urls of all the products

```python
def fetch_urls(url):
    # Getting the Beautiful Soup
    import requests
    from bs4 import BeautifulSoup
    data = requests.get(url)
    bs = BeautifulSoup(data.content)
    # Fetch all product div tags
    divs = bs.find_all('div',class_='_2kHMtA')
    # Fetch the link from each div
    links = []
    for i in divs:
        l = i.find('a').get('href')
        l = 'https://www.flipkart.com'+l
        links.append(l)
    return links    
```

### Create a function to fetch product details

```python
def get_product_details(url):
    # Fetch all the urls
    links = fetch_urls(url)
    # Fetch the product details
    name = []
    price = []
    rating = []
    n_ratings = []
    n_reviews = []
    for i in links:
        data2 = requests.get(i)
        bs2 = BeautifulSoup(data2.content) 
        name.append(bs2.find('span', class_="B_NuCI").text)
        price.append(bs2.find('div', class_="_30jeq3 _16Jk6d").text)
        rating.append(float(bs2.find('div', class_="_3LWZlK").text))
        sp = bs2.find('span',class_="_2_R_DZ").text
        nrat, nrev = sp.split('\xa0&\xa0')
        n_ratings.append(nrat)
        n_reviews.append(nrev)
    # Create a dictionary for the product details
    dct = {'Product Name': name,
           'URL':links,
           'Price':price,
           'Rating':rating,
           'Number of Ratings':n_ratings,
           'Number of Reviews':n_reviews}
    # Convert dictionary to dataframe
    import pandas as pd
    df = pd.DataFrame(dct)
    return df
```

```python
url
```

```python
df_phones = get_product_details(url)
df_phones
```
